### Transforming streaming data

```
dbutils.fs.mkdirs("dbfs:/FileStore/tables/datasets/car_source_stream")
```

```
car_stream_data = spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format", "csv")\
    .option("cloudFiles.schemaLocation",
            "dbfs:/FileStore/tables/datasets/car_source_stream")\
    .option("cloudFiles.schemaHints", "price int, mileage int, engV float, year int")\
    .load("dbfs:/FileStore/tables/datasets/car_source_stream")
```

```
car_stream_data.display()
```

```
car_stream_data.select("year", "price").display()
```

```
car_transform_1 = car_stream_data.select("car", "model", "year", "body", "price")\
    .where("price > 25000")
car_transform_1.display()
```

```
car_transform_2 = car_transform_1.select("*")\
    .filter((car_transform_1["car"].isin(["Mercedes-Benz", "BMW"])) & (car_transform_1["year"] > 2010 ))
car_transform_2.display()
```

```
car_transform_3 = car_transform_2.select("*")\
    .groupBy("year")\
    .count()\
    .withColumnRenamed("count", "total")
car_transform_3.display()
```

```
car_transform_3 = car_transform_2.select("*")\
    .groupBy("year")\
    .count()
car_transform_3.display()
```
